#xml Parsing

```python
def xml_file_list(xml_folder):
  label = glob(xml_folder+'/labelXml'+'/*')
  label = sorted(label, reverse=False)
  label = sorted(label, key=len)
  return label
```

```python
for i in xml_folder_list:
  if 'train_part1' in i:
    train_part1_label = xml_file_list(i)
  elif 'train_part2' in i:
    train_part2_label = xml_file_list(i)
  elif 'validation' in i:
    validation_label = xml_file_list(i)
```

#Detailed Class Frequency Dataframe

```python
def tif_file_detailed_class_frequency_df(label):
  tif_detailed_class_frequency_df = pd.DataFrame()
  
  for i in label:
    tree = ET.parse(i)
    root = tree.getroot()
    detailed_class_df_element = []
    detailed_class_df_element.append(root[0][0].text)
    
    dump_truck = 0; cargo_truck = 0; truck_tractor = 0; tractor = 0; excavator = 0; trailer = 0; van = 0
    small_car = 0; bus = 0; other_vehicle = 0; dry_cargo_ship = 0; liquid_cargo_ship = 0; motorboat = 0
    fishing_boat = 0; tug_boat = 0; engineering_ship = 0; warship = 0; passenger_ship = 0; other_ship = 0
    a220 = 0; a321 = 0; a330 = 0; a350 = 0; boeing737 = 0; boeing747 = 0; boeing777 = 0; boeing787 = 0; arj21 = 0; c919 = 0; other_airplane = 0
    

    for i in root[3]:
      for child in i[3]:
        if 'Dump Truck' == child.text: dump_truck += 1
        if 'Cargo Truck' == child.text: cargo_truck += 1        
        if 'Truck Tractor' == child.text: truck_tractor += 1
        if 'Tractor' == child.text: tractor += 1        
        if 'Excavator' == child.text: excavator += 1        
        if 'Trailer' == child.text: trailer += 1
        if 'Van' == child.text: van += 1  
        if 'Small Car' == child.text: small_car += 1        
        if 'Bus' == child.text: bus += 1
        if 'other-vehicle' == child.text: other_vehicle += 1

        if 'Dry Cargo Ship' == child.text: dry_cargo_ship += 1        
        if 'Liquid Cargo Ship' == child.text: liquid_cargo_ship += 1
        if 'Motorboat' == child.text: motorboat += 1        
        if 'Fishing Boat' == child.text: fishing_boat += 1
        if 'Tugboat' == child.text: tug_boat += 1
        if 'Engineering Ship' == child.text: engineering_ship += 1        
        if 'Warship' == child.text: warship += 1
        if 'Passenger Ship' == child.text: passenger_ship += 1        
        if 'other-ship' == child.text: other_ship += 1 

        if 'A220' == child.text: a220 += 1        
        if 'A321' == child.text: a321 += 1
        if 'A330' == child.text: a330 += 1        
        if 'A350' == child.text: a350 += 1        
        if 'Boeing737' == child.text: boeing737 += 1
        if 'Boeing747' == child.text: boeing747 += 1        
        if 'Boeing777' == child.text: boeing777 += 1
        if 'Boeing787' == child.text: boeing787 += 1
        if 'ARJ21' == child.text: arj21 += 1        
        if 'C919' == child.text: c919 += 1
        if 'other-airplane' == child.text: other_airplane += 1 

    detailed_class_list = [dump_truck, cargo_truck, truck_tractor, tractor, excavator, trailer, 
                           van, small_car, bus, other_vehicle, dry_cargo_ship, liquid_cargo_ship,
                           motorboat, fishing_boat, tug_boat, engineering_ship, warship,
                           passenger_ship, other_ship, a220, a321, a330, a350, boeing737,
                           boeing747, boeing777, boeing787, arj21, c919, other_airplane]

    for i in detailed_class_list:
      detailed_class_df_element.append(i)

    df = pd.DataFrame([detailed_class_df_element])
    tif_detailed_class_frequency_df = tif_detailed_class_frequency_df.append(df)

  tif_detailed_class_frequency_df.columns = ['File', 'Dump Truck', 'Cargo Truck', 'Truck Tractor', 'Tractor', 'Excavator', 'Trailer', 'Van', 'Small Car', 'Bus', 'other-vehicle',
                                             'Dry Cargo Ship', 'Liquid Cargo Ship', 'Motorboat', 'Fishing Boat', 'Tugboat', 'Engineering Ship', 'Warship', 'Passenger Ship', 'other-ship',
                                             'A220', 'A321', 'A330',  'A350', 'Boeing737', 'Boeing747', 'Boeing777', 'Boeing787', 'ARJ21', 'C919', 'other-airplane']

  tif_detailed_class_frequency_df = tif_detailed_class_frequency_df.reset_index()
  del tif_detailed_class_frequency_df['index']
  return tif_detailed_class_frequency_df
```

```python
train_tif_detailed_class_frequency_df = tif_file_detailed_class_frequency_df(train_label)
validation_tif_detailed_class_frequency_df = tif_file_detailed_class_frequency_df(validation_label)
```

#Class Frequency Dataframe

```python
def tif_file_class_frequency_df(label):
  tif_class_frequency_df = pd.DataFrame()
  
  for i in label:
    tree = ET.parse(i)
    root = tree.getroot()
    class_df_element = []
    class_df_element.append(root[0][0].text)

    vehicle = 0
    ship = 0
    airplane = 0

    for i in root[3]:
      for child in i[3]:
        if 'Dump Truck' == child.text or 'Cargo Truck' == child.text or  'Truck Tractor' == child.text \
        or 'Tractor' == child.text or 'Excavator' == child.text or 'Trailer' == child.text or 'Van' == child.text \
        or 'Small Car' == child.text or 'Bus' == child.text or 'other-vehicle' == child.text:
          vehicle += 1
        
        elif 'Dry Cargo Ship' == child.text or 'Liquid Cargo Ship' == child.text or 'Motorboat' == child.text \
        or 'Fishing Boat' == child.text or 'Tugboat' == child.text or 'Engineering Ship' == child.text \
        or 'Warship' == child.text or 'Passenger Ship' == child.text or 'other-ship' == child.text:
          ship += 1

        elif 'A220' == child.text or 'A321' == child.text or 'A330' == child.text or  'A350' == child.text \
        or 'Boeing737' == child.text or 'Boeing747' == child.text or 'Boeing777' == child.text or 'Boeing787' == child.text \
        or 'ARJ21' == child.text or 'C919' == child.text or 'other-airplane' == child.text:
          airplane += 1

    class_list = [vehicle, ship, airplane]

    for i in class_list:
      class_df_element.append(i)

    df = pd.DataFrame([class_df_element])
    tif_class_frequency_df = tif_class_frequency_df.append(df)

  tif_class_frequency_df.columns = ['File', 'Vehicle', 'Ship', 'Airplane']
  tif_class_frequency_df = tif_class_frequency_df.reset_index()
  del tif_class_frequency_df['index']

  return tif_class_frequency_df
```
