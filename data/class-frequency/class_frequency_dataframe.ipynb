#xml Parsing

```python
xml_zip_path  = '/content/drive/MyDrive/Aiffelthon/fair1m2.0'
zip_file_list = [xml_zip_path + '/train/part1/labelXml.zip', xml_zip_path + '/train/part2/labelXmls.zip',
                 xml_zip_path + '/validation/labelXmls.zip']

xml_folder_path = '/content/drive/MyDrive/Aiffelthon/eda/xml_data'
xml_folder_list = [xml_folder_path + '/train/train_part1_xml', xml_folder_path + '/train/train_part2_xml',
                   xml_folder_path + '/validation/vaildation_xml']
```

```python
def extract_xml_zip(zip_path, extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
```

```python
for i in zip_file_list:
  if 'train/part1' in i:
      extract_xml_zip(i, xml_folder_list[0])
  elif 'train/part2' in i:
      extract_xml_zip(i, xml_folder_list[1])
  elif 'validation' in i:
      extract_xml_zip(i, xml_folder_list[2])
```

```python
def xml_file_list(xml_folder):
  label = glob(xml_folder+'/labelXml'+'/*')
  label = sorted(label, reverse=False)
  label = sorted(label, key=len)
  return label
```

```python
for i in xml_folder_list:
  if 'train_part1' in i:
    train_part1_label = xml_file_list(i)
  elif 'train_part2' in i:
    train_part2_label = xml_file_list(i)
  elif 'validation' in i:
    validation_label = xml_file_list(i)
```

```python
train_label = train_part1_label + train_part2_label
```

```python
import xml.etree.ElementTree as ET
import xml.etree.ElementTree

tree = ET.parse(train_label[0])
root = tree.getroot()
ET.dump(tree)
```

```python
print(root[0][0].tag, root[0][0].text)
for i in root[3]:
  for child in i[3]:
    print(child.tag, child.text)
```

#Detailed Class Name Dataframe

```python
def make_detailed_class_name_df(label):
  detailed_class_name_df = pd.DataFrame()
  
  for i in label:
    tree = ET.parse(i)
    root = tree.getroot()
    class_df_element = []
    class_df_element.append(root[0][0].text)
    
    for i in root[3]:
      for child in i[3]:
        class_df_element.append(child.text)
    
    df = pd.DataFrame([class_df_element])
    detailed_class_name_df = detailed_class_name_df.append(df)

  detailed_class_name_df = detailed_class_name_df.reset_index()
  del detailed_class_name_df['index']
  return detailed_class_name_df
```

```python
train_detailed_class_name_df = make_detailed_class_name_df(train_label)
validation_detailed_class_name_df = make_detailed_class_name_df(validation_label)
```

```python
train_detailed_class_name_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_name.csv', index = False)
validation_detailed_class_name_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_detailed_class_name.csv', index = False)
```

#Detailed Class Frequency Dataframe

```python
def make_class_frequency_df(df):
  df = df.stack().value_counts().rename_axis('value').reset_index(name='count')
  xml_class_frequency_df = df[~df['value'].str.contains("tif", na=False, case=False)]
  return xml_class_frequency_df
```

```python
train_detailed_class_frequency_df = make_class_frequency_df(train_detailed_class_name_df)
validation_detailed_class_frequency_df = make_class_frequency_df(validation_detailed_class_name_df)
```

```python
train_detailed_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_frequency.csv', index = False)
validation_detailed_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_detailed_class_frequency.csv', index = False)
```

#Class Frequency Dataframe

```python
train_detailed_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_frequency.csv')
validation_detailed_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_detailed_class_frequency.csv')
```

```python
print("Detailed Class")
print()
count = 0
for key, value in train_detailed_class_frequency_df['value'].to_dict().items():
  count += 1
  if count == len(train_detailed_class_frequency_df):
    print(value, end = ' ')
  elif count < len(train_detailed_class_frequency_df):
    print(value, end = ', ')
    if count % 9 == 0 and count <= 27:
      print()
```

```python
class_list = ["Vehicle", "Ship", "Airplane"]
print("Class:", ", ".join(class_list))
```

```python
vehicle_class_list = ['Dump Truck', 'Cargo Truck', 'Truck Tractor', 'Tractor','Excavator', 'Trailer', 'Van', 'Small Car', 'Bus', 'other-vehicle']
train_vehicle_class_frequency_df = train_detailed_class_frequency_df[train_detailed_class_frequency_df['value'].isin(vehicle_class_list)].sort_values(by='count', ascending=False).reset_index()
del train_vehicle_class_frequency_df['index']

validation_vehicle_class_frequency_df = validation_detailed_class_frequency_df[validation_detailed_class_frequency_df['value'].isin(vehicle_class_list)].sort_values(by='count', ascending=False).reset_index()
del validation_vehicle_class_frequency_df['index']
```

```python
ship_class_list = ['Dry Cargo Ship', 'Liquid Cargo Ship', 'Motorboat', 'Fishing Boat', 'Tugboat', 'Engineering Ship', 'Warship', 'Passenger Ship', 'other-ship']
train_ship_class_frequency_df = train_detailed_class_frequency_df[train_detailed_class_frequency_df['value'].isin(ship_class_list)].sort_values(by='count', ascending=False).reset_index()
del train_ship_class_frequency_df['index']

validation_ship_class_frequency_df = validation_detailed_class_frequency_df[validation_detailed_class_frequency_df['value'].isin(ship_class_list)].sort_values(by='count', ascending=False).reset_index()
del validation_ship_class_frequency_df['index']
```

```python
airplane_class_list = ['A220', 'A321', 'A330',  'A350', 'Boeing737', 'Boeing747', 'Boeing777', 'Boeing787', 'ARJ21', 'C919', 'other-airplane']
train_airplane_class_frequency_df = train_detailed_class_frequency_df[train_detailed_class_frequency_df['value'].isin(airplane_class_list)].sort_values(by='count', ascending=False).reset_index()
del train_airplane_class_frequency_df['index']

validation_airplane_class_frequency_df = validation_detailed_class_frequency_df[validation_detailed_class_frequency_df['value'].isin(airplane_class_list)].sort_values(by='count', ascending=False).reset_index()
del validation_airplane_class_frequency_df['index']
```

```python
train_vehicle_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_vehicle_class_frequency.csv', index = False)
train_ship_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_ship_class_frequency.csv', index = False)
train_airplane_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_airplane_class_frequency.csv', index = False)

validation_vehicle_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_vehicle_class_frequency.csv', index = False)
validation_ship_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_ship_class_frequency.csv', index = False)
validation_airplane_class_frequency_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_airplane_class_frequency.csv', index = False)
```
