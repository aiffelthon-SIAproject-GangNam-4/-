#xml Parsing

```python
def xml_file_list(xml_folder):
  label = glob(xml_folder+'/labelXml'+'/*')
  label = sorted(label, reverse=False)
  label = sorted(label, key=len)
  return label
```

```python
xml_folder_path = '/content/drive/MyDrive/Aiffelthon/eda/xml_data'
xml_folder_list = [xml_folder_path + '/train/train_part1_xml', xml_folder_path + '/train/train_part2_xml',
                   xml_folder_path + '/validation/vaildation_xml']
```

```python
for i in xml_folder_list:
  if 'train_part1' in i:
    train_part1_label = xml_file_list(i)
  elif 'train_part2' in i:
    train_part2_label = xml_file_list(i)
  elif 'validation' in i:
    validation_label = xml_file_list(i)
```

```python
def namestr(obj, namespace):
    return [name for name in namespace if namespace[name] is obj]
```

```python
def save_detailed_tif_file_name_csv(df_name, tif_file_list, number):
  globals()[df_name] = pd.DataFrame({'file':tif_file_list[number]})
  globals()[df_name].to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_tif_file_name_path/' + df_name.split("_")[0] +'_detailed_class_tif_file_name/' + df_name + '.csv', index = False)
```

```python
def save_detailed_file_name_count_csv(folder_part):
  class_name_list = []
  tif_file_count_list = []
  
  csv_files = glob('/content/drive/MyDrive/Aiffelthon/eda/class_tif_file_name_path/' + folder_part + 'detailed_class_tif_file_name/' + '*.csv')
  
  for f in csv_files:
    df = pd.read_csv(f)
    class_name = ' '.join(s for s in f.split("/")[-1][:-4].split("_")[1:-3])
    class_name_list.append(class_name)
    tif_file_count = len(df)
    tif_file_count_list.append(tif_file_count)

  file_count_df = pd.DataFrame({'value':class_name_list, 'count':tif_file_count_list})

  reorder_list = ['dump truck', 'cargo truck', 'truck tractor', 'tractor', 'excavator', 'trailer', 
                  'van', 'small car', 'bus', 'other vehicle', 'dry cargo ship', 'liquid cargo ship',
                  'motorboat', 'fishing boat', 'tug boat', 'engineering ship', 'warship',
                  'passenger ship', 'other ship', 'a220', 'a321', 'a330', 'a350', 'boeing737',
                  'boeing747', 'boeing777', 'boeing787', 'arj21', 'c919', 'other airplane']

  sorterIndex = dict(zip(reorder_list, range(len(reorder_list))))
  file_count_df['order'] = file_count_df['value'].map(sorterIndex)
  file_count_df.sort_values(['order', 'value', 'count'], ascending = [True, True, True], inplace = True)
  file_count_df.drop('order', axis = 1, inplace = True)

  file_count_df = file_count_df.reset_index()
  del file_count_df['index']
  file_count_df = file_count_df[:-1]

  file_count_df.to_csv('/content/drive/MyDrive/Aiffelthon/eda/class_tif_file_name_path/' + folder_part + 'detailed_class_tif_file_name/' + 'file_count_df.csv', index = False)
```

```python
def detailed_class_tif_file_name(label, class_list, tif_file_list):
  
  for i in label:
    tree = ET.parse(i)
    root = tree.getroot()
    for j in class_list:
      list_index = class_list.index(j)         
      for k in root[3]:
        for child in k[3]:
          if child.text == class_list[list_index]:
            if root[0][0].text not in tif_file_list[list_index]:
              tif_file_list[list_index].append(root[0][0].text)

  for i in range(len(tif_file_list)):
    df_name = [k for k,v in globals().items() if id(v) == id(tif_file_list[i])][0].replace('list', 'df')
    if 'train' in namestr(label, globals())[0]:
      df_name = 'train_' + df_name 
    elif 'validation' in namestr(label, globals())[0]:
      df_name = 'validation_' + df_name 
    save_detailed_tif_file_name_csv(df_name, tif_file_list, i)

  if 'train' in namestr(label, globals())[0]:
    save_detailed_file_name_count_csv('train_')
  elif 'validation' in namestr(label, globals())[0]:
    save_detailed_file_name_count_csv('validation_')
```
