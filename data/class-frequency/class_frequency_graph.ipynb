```python
import pandas as pd
from termcolor import colored
import matplotlib.pyplot as plt
from IPython.display import HTML
from IPython.display import display
```

```python
train_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_class_frequency.csv')
validation_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_class_frequency.csv')
```

```python
train_vehicle_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_vehicle_class_frequency.csv')
train_ship_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_ship_class_frequency.csv')
train_airplane_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_airplane_class_frequency.csv')
validation_vehicle_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_vehicle_class_frequency.csv')
validation_ship_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_ship_class_frequency.csv')
validation_airplane_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_airplane_class_frequency.csv')
```

```python
xs = train_class_frequency_df.value.to_list()
ys = train_class_frequency_df['count'].to_list()

print(colored("Train Data", attrs=['bold']))
print()
plt.figure(figsize=(8, 5))
plt.title('Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_class_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
print("\n"*2)

xs = validation_class_frequency_df.value.to_list()
ys = validation_class_frequency_df['count'].to_list()

print(colored("Validation Data", attrs=['bold']))
print()
plt.figure(figsize=(8, 5))
plt.title('Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_class_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
```

#Detailed Class Frequency Graph

```python
train_detailed_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_frequency.csv')
validation_detailed_class_frequency_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/validation_detailed_class_frequency.csv')
```

```python
xs = train_detailed_class_frequency_df.value.to_list()
ys = train_detailed_class_frequency_df['count'].to_list()

print(colored("Train Data", attrs=['bold']))
print()
plt.figure(figsize=(60, 11))
plt.title('Detailed Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
print("\n"*2)

xs = train_detailed_class_frequency_df.value.to_list()[:10]
ys = train_detailed_class_frequency_df['count'].to_list()[:10]

plt.figure(figsize=(25, 10))
plt.title('Detailed Class 1-10')
plt.xlabel('Class')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_1-10_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
print("\n"*2)

xs = train_detailed_class_frequency_df.value.to_list()[3:20]
ys = train_detailed_class_frequency_df['count'].to_list()[3:20]

plt.figure(figsize=(27, 10))
plt.title('Detailed Class 4-20')
plt.xlabel('Class')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_4-20_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
print("\n"*2)

xs = train_detailed_class_frequency_df.value.to_list()[20:]
ys = train_detailed_class_frequency_df['count'].to_list()[20:]

plt.figure(figsize=(30, 10))
plt.title('Detailed Class 21-37')
plt.xlabel('Class')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/class_frequency/train_detailed_class_21-37_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
```
