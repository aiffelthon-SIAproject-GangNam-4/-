```python
import pandas as pd
from pandas import option_context
from termcolor import colored
import matplotlib.pyplot as plt
from IPython.display import HTML
from IPython.display import display
```

```python
train_image_size_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/image_size/df/train_image_size_df.csv')
validation_image_size_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/image_size/df/validation_image_size_df.csv')
```

```python
train_image_size_count_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/image_size/df/train_image_size_count_df.csv')
validation_image_size_count_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/image_size/df/validation_image_size_count_df.csv')
```

```python
train_image_size_count_sort_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/image_size/df/train_image_size_count_sort_df.csv')
validation_image_size_count_sort_df = pd.read_csv('/content/drive/MyDrive/Aiffelthon/eda/image_size/df/validation_image_size_count_sort_df.csv')
```

```python
print(colored("Train Data", attrs=['bold']))
print()
print(train_image_size_count_df['size'].unique())
print("\n")

print(colored("Validation Data", attrs=['bold']))
print()
print(validation_image_size_count_df['size'].unique())
```

```python
print(colored("Train Data", attrs=['bold']))
print()
print(train_image_size_count_sort_df['size'].unique())
print("\n")

print(colored("Validation Data", attrs=['bold']))
print()
print(validation_image_size_count_sort_df['size'].unique())
```

```python
xs = train_image_size_count_df['size'].to_list()[:10]
ys = train_image_size_count_df['count'].to_list()[:10]

print(colored("Train Data", attrs=['bold']))
print()
plt.figure(figsize=(30, 5))
plt.title('Image Size 1-10')
plt.xlabel('Size')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/image_size/graph/train_image_size_count_1-10_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()

print("\n"*2)

xs = train_image_size_count_df['size'].to_list()[10:20]
ys = train_image_size_count_df['count'].to_list()[10:20]

plt.figure(figsize=(30, 5))
plt.title('Image Size 11-20')
plt.xlabel('Size')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/image_size/graph/train_image_size_count_11-20_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()

print("\n"*2)

xs = train_image_size_count_df['size'].to_list()[21:]
ys = train_image_size_count_df['count'].to_list()[21:]

plt.figure(figsize=(30, 5))
plt.title('Image Size 21-36')
plt.xlabel('Size')
plt.ylabel('Count')
plt.bar(xs, ys, width=0.8, color='gray')
plt.savefig('/content/drive/MyDrive/Aiffelthon/eda/image_size/graph/train_image_size_count_21-36_graph.png',
            facecolor='#eeeeee', edgecolor='black', format='png', bbox_inches='tight')
plt.show()
```
